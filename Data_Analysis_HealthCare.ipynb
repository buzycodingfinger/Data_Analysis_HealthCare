#Analysis based on gender and heart disease.

1 = male and 0 = female

```python
female_heartDis=data[(data['sex']==0) & (data['target']==1)]
female_Count=female_heartDis['target'].sum()
female_Count
```

```python
male_heartDis=data[(data['sex']==1) & (data['target']==1)]
male_Count=male_heartDis['target'].sum()
male_Count
```

```python
sns.catplot(x='sex',y='target',kind='boxen',data=data)
```

Conclusion: Males have more heart disease than Females.

Here are the ranges for total cholesterol in adults:

 Normal: Less than 200 mg/dL.

 Borderline high: 200 to 239 mg/dL.

 Treatment: Check the range of cholesterol and follow a healty diet and prevent Heart disease.
